# Exporting 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x = np.linspace(0,np.pi,9)
```

```python
y = np.sin(x)
```

```python
f = plt.figure(dpi=140)
ax = f.add_axes([0,0,1,1])
ax.plot(x,y,label="X,Y")
ax.plot(y,x,label="Y,X")
ax.set_title("Label")
ax.set_xlabel("xLabel")
ax.set_ylabel("yLabel")
ax.legend(loc=(.1,.1))
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```

```python
fig, axes = plt.subplots(dpi=200)
axes.plot(y,x,color="cyan",linewidth=1, marker="^", markeredgewidth=5, 
          markeredgecolor="green",markerfacecolor="purple",linestyle="-")
```

```python
fig.savefig('1st Plot.png',dpi=50)
```
