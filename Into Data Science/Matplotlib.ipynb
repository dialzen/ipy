# The OO Method

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.7,0.7]) # Left, Bottom, Width, Hieght
axes.set_xlabel("X Label")
axes.plot(x,y)
axes2 = fig.add_axes([0.1,0.1,0.3,0.3])
axes2.plot(x,y)
axes2.set_xlabel("X2 label")
```

```python
newfig = plt.figure()
axes1 = newfig.add_axes([0,0,1,1])
axes2 = newfig.add_axes([0.1,0,1,1])
axes1.plot(x,y)
axes2.plot(x,y,'r')
```
