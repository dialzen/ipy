```
import numpy as np


A = np.matrix([[1,1,-1],[0,1,3],[0,0,-6]])
b = np.array([9,3,8],dtype=float)


def back_sub(A,b):
    n = len(A)
    print('n is:', n)
    x = [0]*n
    for i in range(n-1,-1,-1): #this refers to the rows; i goes 2,1,0
        for j in range(i+1,n): #j goes 1,2 @ i = 0
                               #j goes 2   @ i = 1
            b[i] = b[i] - A[i,j]*x[j]
        x[i] = b[i]/A[i,i]

    return x
```

```
back_sub(A,b)
```
